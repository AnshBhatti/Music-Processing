```python
import librosa as lib
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
```

```python
def get_data(genre,n):
    audio_data="C:/Users/Ansh/Desktop/MusicClassification/genres/"+genre+'/'+genre+'.'+'0'*(5-len(str(n)))+str(n)+".wav"                                                                                        
    x,sr=lib.load(audio_data)
    x=x[:330000]
    data=np.array([x])
    data=torch.tensor(data)
    #data=data.view(-1,1)
    return data
get_data("blues",10)
```

```python
model=nn.Sequential(nn.Linear(330000,100),nn.Linear(100,10),nn.LogSoftmax(dim=1))
optimizer=optim.SGD(model.parameters(),lr=0.01,momentum=0.7)
criterion=nn.NLLLoss()
arr="blues classical country disco hiphop jazz metal pop reggae rock".split()
```

```python
for epoch in range(10):
    for a in range(10):
        for i in range(50):
            data=get_data(arr[a],i)
            #print(data)
            out=model(data)
            #print(out)
            ar=torch.tensor([a])
            loss=criterion(out,ar)
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()
    print(epoch)
```

```python
data=get_data("blues",70)
torch.exp(model(data))
```
